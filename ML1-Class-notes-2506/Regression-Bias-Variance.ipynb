# Section B : Auto CV Dataset

```python
auto = pd.read_csv("C:\\Users\\PC\\Machine-Learning-class\\DATA\\auto (1).csv")
```

```python
# Drop the "origin column"

auto = auto.drop("origin",axis=1)
```

```python
x = auto.drop("mpg",axis=1)
y = auto["mpg"]
```

## Making a Prediction with a Single Feature


```python
X_hp = x[["hp"]]
X_hp.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_hp["hp"], y)
plt.ylabel("Miles Per Gallon(mpg)")
plt.xlabel("Horsepower (hp)")
plt.show()
```

```python
reg = LinearRegression()
reg.fit(X_hp,y)
predictions = reg.predict(X_hp)
import matplotlib.pyplot as plt
plt.scatter(X_hp,y)
plt.plot(X_hp,predictions,color="red")
plt.ylabel("Milesper Gallon (mpg)")
plt.xlabel('Horse power (hp)')
plt.show()
```

```python

Xhp_train, Xhp_test, y_train, y_test = train_test_split(X_hp, y, test_size=0.2, random_state=42)

reg2 = LinearRegression()
reg2.fit(Xhp_train, y_train)

predictions2 = reg2.predict(Xhp_test)

plt.scatter(Xhp_test, y_test)
plt.plot(Xhp_test, predictions2, color='red')
plt.ylabel('Miles Per Gallon (mpg)')
plt.xlabel('Horsepower (hp)')
plt.show()
```
