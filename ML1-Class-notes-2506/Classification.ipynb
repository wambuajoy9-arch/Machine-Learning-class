## Classification metrics
True Positive(Tp)- the model says that these are defaulters and they are defaulters
False Positive -Model says they are defaulters and they are not
True Negative-model says that they are non defaulters and they are
False Negative-vice versa


1. ACCURACY:
- Out of all the predictions the model was supposed to do how many times was it correct?
**Correct Predictions/All predictions**

2. PRECISION:
- Out of all the defaults in the model how many where the actual defaulters?
**TP/(TP+FP)**
- Used when you want to reduce false positives

3.RECALL:
- Out of all the defaulters how many did  the model actually catch -"recall"
- Used to minimize fals negatives
**TP/(FN+TP)**
