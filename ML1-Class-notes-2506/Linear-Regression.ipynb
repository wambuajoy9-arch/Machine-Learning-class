### Bias-Variance Tradeoff
- Bias is the difference between the predicted and actual value
- This is an error caused by overly simple assumptions in the model which may lead to underfitting
- If a model is high biased it is underfitting then you should make a complex model as the fix
- Variance is how the models prediction changes when its fed different input... A high variance model is overfitting
- It is an error caused by the model being too senstive to training data, which may lead to overfitting then we should make a simple model
- A good machine model is one that can balance bias and variance
- A high bias and variance model is identified using learning curves
- If the curves are both low then that is a high biased model


## Learning curve
- It is a graphical represesntation that shows a models perfomance changes as it learns from more data or training iterations .
- It helps evaluate whether a model is learning effectively and can be used to identify underfitting,overfitting and well fitted model.
- **Training Curve**- Perfomance of the model on training data
- **Validation curve**- Perfomance of the model on unseen validation data
- **X-axis**- Amount of training data or training iterations
- **Y-axis**- Perfomance metric such as accuracy,error or loss

## Identifying model behaviour using learning curves
1. Well-Fitted Model
- it learns the underlying patterns without memorizing the training data:
 - training and validation curves are close to each other
 - both curves achieve good perfomance
 - small gap between the training and validation error
 - good generalization to unseen data


2. Overfitting
Overfitting occurs when a model learns the training data too well, including noise and irrelevant patterns. As a result, it performs well on training data but poorly on unseen data.

- Training error is very low.
- Validation error remains high.
- Large gap between training and validation curves.
- Poor generalization on new data.


3. Underfitting
Underfitting occurs when the model is too simple to capture the underlying patterns in the data. It performs poorly on both training and validation datasets.

- High training error.
- High validation error.
- Both curves remain close but at poor performance levels.
- Model fails to learn important patterns.

#### Regularization
- How do we address Overfitting?
1. Adding more training data(The more data you add it becomes less esnsitive to the noise even with the same high order polynomial)

2. Feature selection
- the more complex features you have the higher risk of overfitting
- the more features you have the more dimensions you have and the more complex the feature  is
- Dropping features help


3. Regularization(used when a model is overfitting)
- It is a way to penalize complex models
- A regularized model has two priorities
1. minimize the loss function
2. Keep the coefficients small

**Regularization Techniques**
1. Lasso Regularization(L1)(sum of absolute values)
- Penalizes the absolute values for each coefficient
- Pushes the coeeficients to zero- auto feature selection
-Not very good for multicollinearity- it will pick the one and ignore the rest 
- Produces a sparse model-model with fewer features


2. Ridge Regularization(L2)(sum of squares)
-Penalizes the squared values of each coefficient

3. Elastic Net regularization(sum of absolute values and squares)
- Penalizes the absolute valuesof each coefficient instead


When is regularization used?
- When you have high multicollinearity use **Ridge or Elastic Net**
- When you want a sparse model use **Lasso**
- When you have irrelevant features **lasso**
- when you have a stable model **ridge**
